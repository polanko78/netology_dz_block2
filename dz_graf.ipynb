```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_res = pd.DataFrame()
for dt in range(1900, 2001):
    name = 'names\yob' + str(dt) +'.txt'
    df = pd.read_csv(name, names = ['Name', 'Gender', 'Counts'])
    df['Year'] = dt
    if df_res.empty:
        df_res = df
    else:
        df_res = pd.concat([df_res, df])
ruth = df_res[df_res['Name'] == 'Ruth'].groupby(['Year']).sum()
robert = df_res[df_res['Name'] == 'Robert'].groupby(['Year']).sum()
plt.plot(robert)
plt.plot(ruth)
```

```python
def foo(df):
    for x in df.index:
        if x%5 == 0:
            pass
        else:
            df = df.drop(x)
    return df
res_ruth = foo(ruth)
res_robert = foo(robert)
res_ruth.plot(kind = 'bar')
res_robert.plot(kind = 'bar')
```

```python
name = 'names\yob1950.txt'
df = pd.read_csv(name, names = ['Name', 'Gender', 'Counts'])
df['Format'] = df.apply(lambda x: 'R' in x['Name'], axis = 1)
names = df[df['Format'] == True].groupby(['Name']).sum().sort_values('Counts', ascending = False)[:10]
names.plot(kind='pie', y='Counts')
```

```python
def count(df_res):
    res1 = pd.DataFrame()
    letter_list = ['q', 'w', 'r', 't', 'p', 's', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'z', 'x', 'c', 'v', 'b', 'n', 'm']
    for x in df_res.index:
        counter = 0
        for letter in x.lower():
            if letter in letter_list:
                counter += 1
        my_dict = {'Name': x, 'Number': counter, 'Counts': df_res.loc[x, 'Counts']}
        res = pd.DataFrame([my_dict])
        res1 = pd.concat([res1, res])
    return res1
    
df_res = df_res.groupby(['Name']).sum()
#df_res['Number'] = df_res.apply(count, axis = 1)
res1 = count(df_res)
```

```python
res1.plot(kind='scatter', x='Number', y='Counts')
```
